```python
#!/usr/bin/env python3
"""
Revised game-theoretic electromagnetic follow-up scheduler for public gravitational-wave sky maps.

This script implements a reproducible pipeline for public GW sky maps using:
- gwosc        : event metadata (GPS time, detectors)
- ligo.skymap  : reading HEALPix/FITS sky maps
- astroplan    : visibility and night-time feasibility

The methodology mirrors the manuscript draft:
1. Public-event ingestion.
2. Candidate tile generation over the high-posterior region.
3. Telescope/time feasibility generation.
4. Proposed method: potential-game equilibrium via exact local best responses
   on a pruned action library.
5. Benchmarks: random feasible, independent greedy, centralized greedy.
6. Tables, plots, and ZIP packaging of all outputs.

Designed for Google Colab or a local Python environment.
"""

from __future__ import annotations

import argparse
import itertools
import json
import math
import os
import random
import textwrap
import warnings
import zipfile
from dataclasses import dataclass, asdict
from pathlib import Path
from typing import Dict, Iterable, List, Optional, Sequence, Tuple

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import astropy.units as u
from astropy.coordinates import AltAz, EarthLocation, SkyCoord, get_sun
from astropy.time import Time

import healpy as hp
import requests
from gwosc.datasets import event_detectors, event_gps
from ligo.skymap.io.fits import read_sky_map
from astroplan import Observer, download_IERS_A

try:
    from IPython.display import display
except Exception:  # pragma: no cover
    display = None


# -----------------------------------------------------------------------------
# Configuration
# -----------------------------------------------------------------------------

PUBLIC_EVENT_CONFIGS = {
    "GW170817": {
        "skymap_url": "https://dcc.ligo.org/public/0146/G1701985/001/LALInference_v2.fits.gz",
        "catalog_note": "GWOSC public event page + LIGO DCC publication localization",
    },
    "GW190814": {
        "skymap_url": "https://dcc.ligo.org/public/0169/P2000230/005/GW190814_skymap.fits.gz",
        "catalog_note": "GWOSC event version page + LIGO DCC skymap release",
    },
}

# Synthetic but astronomically plausible telescope network for methodology evaluation.
# Coordinates are real observing sites; FoV/depth parameters are illustrative.
TELESCOPE_CONFIGS = [
    {
        "name": "Network_A_ZTFLike",
        "site_label": "Palomar-like",
        "lat_deg": 33.3563,
        "lon_deg": -116.8650,
        "height_m": 1712.0,
        "fov_sqdeg": 47.0,
        "depth_weight": 0.82,
        "exposure_s": 300.0,
        "overhead_s": 45.0,
        "max_actions": 3,
        "min_alt_deg": 25.0,
    },
    {
        "name": "Network_B_DECamLike",
        "site_label": "CTIO-like",
        "lat_deg": -30.1690,
        "lon_deg": -70.8040,
        "height_m": 2200.0,
        "fov_sqdeg": 3.0,
        "depth_weight": 1.00,
        "exposure_s": 420.0,
        "overhead_s": 60.0,
        "max_actions": 3,
        "min_alt_deg": 25.0,
    },
    {
        "name": "Network_C_MeerLICHTLike",
        "site_label": "SAAO-like",
        "lat_deg": -32.3794,
        "lon_deg": 20.8107,
        "height_m": 1798.0,
        "fov_sqdeg": 2.7,
        "depth_weight": 0.93,
        "exposure_s": 300.0,
        "overhead_s": 40.0,
        "max_actions": 3,
        "min_alt_deg": 25.0,
    },
    {
        "name": "Network_D_SkyMapperLike",
        "site_label": "SidingSpring-like",
        "lat_deg": -31.2733,
        "lon_deg": 149.0644,
        "height_m": 1165.0,
        "fov_sqdeg": 5.7,
        "depth_weight": 0.88,
        "exposure_s": 320.0,
        "overhead_s": 45.0,
        "max_actions": 3,
        "min_alt_deg": 25.0,
    },
    {
        "name": "Network_E_BlackGEMLike",
        "site_label": "LaSilla-like",
        "lat_deg": -29.2613,
        "lon_deg": -70.7346,
        "height_m": 2400.0,
        "fov_sqdeg": 2.7,
        "depth_weight": 0.96,
        "exposure_s": 300.0,
        "overhead_s": 35.0,
        "max_actions": 3,
        "min_alt_deg": 25.0,
    },
]

METHOD_ORDER = [
    "proposed_potential_game",
    "centralized_greedy",
    "independent_greedy",
    "random_feasible",
]

METHOD_LABELS = {
    "proposed_potential_game": "Proposed potential game",
    "centralized_greedy": "Centralized greedy",
    "independent_greedy": "Selfish independent greedy",
    "random_feasible": "Random feasible",
}

METHOD_COLORS = {
    "proposed_potential_game": "tab:blue",
    "centralized_greedy": "tab:green",
    "independent_greedy": "tab:orange",
    "random_feasible": "tab:red",
}


# -----------------------------------------------------------------------------
# Data classes
# -----------------------------------------------------------------------------

@dataclass(frozen=True)
class Telescope:
    name: str
    site_label: str
    lat_deg: float
    lon_deg: float
    height_m: float
    fov_sqdeg: float
    depth_weight: float
    exposure_s: float
    overhead_s: float
    max_actions: int
    min_alt_deg: float

    @property
    def radius_deg(self) -> float:
        return float(np.sqrt(self.fov_sqdeg / np.pi))

    @property
    def observer(self) -> Observer:
        location = EarthLocation(
            lat=self.lat_deg * u.deg,
            lon=self.lon_deg * u.deg,
            height=self.height_m * u.m,
        )
        return Observer(location=location, name=self.name, timezone="UTC")


@dataclass(frozen=True)
class Tile:
    tile_id: int
    ra_deg: float
    dec_deg: float
    center_prob: float


@dataclass(frozen=True)
class Action:
    action_id: int
    telescope: str
    tile_id: int
    slot_id: int
    slot_time_iso: str
    ra_deg: float
    dec_deg: float
    radius_deg: float
    q_scalar: float
    cost: float
    slot_hours_since_event: float
    visibility_score: float
    depth_weight: float
    covered_pixels: Tuple[int, ...]


# -----------------------------------------------------------------------------
# Helper utilities
# -----------------------------------------------------------------------------

def safe_mkdir(path: Path) -> None:
    path.mkdir(parents=True, exist_ok=True)


def maybe_display_dataframe(df: pd.DataFrame, title: str) -> None:
    print("\n" + "=" * 100)
    print(title)
    print("=" * 100)
    with pd.option_context("display.max_rows", 200, "display.max_columns", 50, "display.width", 200):
        print(df)
    if display is not None:
        try:
            display(df)
        except Exception:
            pass


def set_seed(seed: int) -> None:
    random.seed(seed)
    np.random.seed(seed)


def ensure_iers() -> None:
    try:
        download_IERS_A(show_progress=False)
    except Exception as exc:
        warnings.warn(f"IERS_A download failed; continuing with cached/default Earth rotation data. Details: {exc}")


def download_file(url: str, destination: Path, timeout: int = 60) -> Path:
    destination.parent.mkdir(parents=True, exist_ok=True)
    if destination.exists() and destination.stat().st_size > 0:
        return destination
    print(f"Downloading: {url}")
    with requests.get(url, stream=True, timeout=timeout) as response:
        response.raise_for_status()
        with open(destination, "wb") as f:
            for chunk in response.iter_content(chunk_size=2**20):
                if chunk:
                    f.write(chunk)
    return destination


def credible_levels(prob: np.ndarray) -> np.ndarray:
    order = np.argsort(prob)[::-1]
    cdf = np.cumsum(prob[order])
    levels = np.empty_like(prob)
    levels[order] = cdf
    return levels


def compute_area_sqdeg(mask: np.ndarray, nside: int) -> float:
    return float(mask.sum() * hp.nside2pixarea(nside, degrees=True))


def hp_pix_to_skycoord(nside: int, pix: np.ndarray) -> SkyCoord:
    theta, phi = hp.pix2ang(nside, pix)
    ra = np.degrees(phi)
    dec = 90.0 - np.degrees(theta)
    return SkyCoord(ra=ra * u.deg, dec=dec * u.deg, frame="icrs")


def skycoord_to_vec(coord: SkyCoord) -> np.ndarray:
    return hp.ang2vec(0.5 * np.pi - coord.dec.radian, coord.ra.radian)


def wrap_ra_to_mollweide(ra_deg: np.ndarray) -> np.ndarray:
    ra_rad = np.radians(ra_deg)
    wrapped = np.remainder(ra_rad + np.pi, 2 * np.pi) - np.pi
    return -wrapped


def generate_time_slots(event_time: Time, latency_hours: float, horizon_hours: float, slot_minutes: int) -> List[Time]:
    n_steps = int(np.floor((horizon_hours * 60.0) / slot_minutes)) + 1
    return [event_time + (latency_hours + i * slot_minutes / 60.0) * u.hour for i in range(n_steps)]


# -----------------------------------------------------------------------------
# Event ingestion and skymap processing
# -----------------------------------------------------------------------------

def load_public_event_skymap(
    event_name: str,
    cache_dir: Path,
    working_nside: int,
) -> Dict[str, object]:
    if event_name not in PUBLIC_EVENT_CONFIGS:
        raise ValueError(f"Event {event_name!r} is not in PUBLIC_EVENT_CONFIGS. Add a verified public skymap URL first.")

    event_cfg = PUBLIC_EVENT_CONFIGS[event_name]
    skymap_url = event_cfg["skymap_url"]
    local_path = cache_dir / f"{event_name}.fits.gz"
    download_file(skymap_url, local_path)

    prob, meta = read_sky_map(str(local_path), nest=False)
    prob = np.asarray(prob, dtype=float)
    prob = np.nan_to_num(prob, nan=0.0, posinf=0.0, neginf=0.0)
    prob = np.clip(prob, 0.0, None)

    input_nside = hp.get_nside(prob)
    if input_nside != working_nside:
        prob = hp.ud_grade(prob, nside_out=working_nside, power=-2)
        prob = np.clip(prob, 0.0, None)

    if prob.sum() <= 0:
        raise ValueError(f"Skymap for {event_name} has non-positive total probability after preprocessing.")
    prob = prob / prob.sum()
    nside = hp.get_nside(prob)

    try:
        gps = float(event_gps(event_name))
    except Exception:
        gps = float(meta.get("gps_time", np.nan))

    try:
        detectors = sorted(list(event_detectors(event_name)))
    except Exception:
        detectors = []

    event_time = Time(gps, format="gps", scale="utc") if np.isfinite(gps) else None

    cl = credible_levels(prob)
    area50 = compute_area_sqdeg(cl <= 0.50, nside)
    area90 = compute_area_sqdeg(cl <= 0.90, nside)
    area99 = compute_area_sqdeg(cl <= 0.99, nside)

    return {
        "event_name": event_name,
        "skymap_url": skymap_url,
        "local_path": str(local_path),
        "prob": prob,
        "meta": meta,
        "gps": gps,
        "event_time": event_time,
        "detectors": detectors,
        "nside": nside,
        "pixel_area_sqdeg": hp.nside2pixarea(nside, degrees=True),
        "credible_levels": cl,
        "area50_sqdeg": area50,
        "area90_sqdeg": area90,
        "area99_sqdeg": area99,
    }


# -----------------------------------------------------------------------------
# Tile generation
# -----------------------------------------------------------------------------

def choose_tile_centers(
    prob: np.ndarray,
    nside: int,
    max_tiles: int = 20,
    candidate_credible_level: float = 0.99,
    min_separation_deg: float = 2.0,
    selection_radius_deg: float = 2.0,
) -> List[Tile]:
    """
    Greedy tile seeding using residual posterior mass inside a disc.
    This is intentionally more diverse than ranking by pixel center probability alone.
    """
    cl = credible_levels(prob)
    candidate_pixels = np.where(cl <= candidate_credible_level)[0]
    residual = prob.copy()
    selected: List[Tile] = []
    selected_coords: List[SkyCoord] = []

    for _ in range(max_tiles):
        best_pix = None
        best_mass = -1.0
        for pix in candidate_pixels:
            if residual[pix] <= 0:
                continue
            coord = hp_pix_to_skycoord(nside, np.array([pix]))[0]
            if selected_coords:
                sep = np.min(coord.separation(SkyCoord(selected_coords)).deg)
                if sep < min_separation_deg:
                    continue
            vec = skycoord_to_vec(coord)
            disc = hp.query_disc(nside, vec, np.radians(selection_radius_deg), inclusive=False, nest=False)
            mass = float(residual[disc].sum())
            if mass > best_mass + 1e-15:
                best_mass = mass
                best_pix = int(pix)

        if best_pix is None or best_mass <= 0:
            break

        coord = hp_pix_to_skycoord(nside, np.array([best_pix]))[0]
        vec = skycoord_to_vec(coord)
        disc = hp.query_disc(nside, vec, np.radians(selection_radius_deg), inclusive=False, nest=False)
        selected.append(
            Tile(
                tile_id=len(selected),
                ra_deg=float(coord.ra.deg),
                dec_deg=float(coord.dec.deg),
                center_prob=float(prob[best_pix]),
            )
        )
        selected_coords.append(coord)
        residual[disc] = 0.0

    return selected


# -----------------------------------------------------------------------------
# Feasible action generation
# -----------------------------------------------------------------------------

def altitude_visibility_score(alt_deg: float, min_alt_deg: float) -> float:
    if alt_deg < min_alt_deg:
        return 0.0
    # Smoothly rescale [min_alt, 90] -> [0, 1].
    score = (alt_deg - min_alt_deg) / max(1e-6, 90.0 - min_alt_deg)
    return float(np.clip(score, 0.0, 1.0))


def darkness_score(sun_alt_deg: float) -> float:
    # Fully dark at -18 deg or lower; linearly downweight twilight between -18 and -8 deg.
    if sun_alt_deg <= -18.0:
        return 1.0
    if sun_alt_deg >= -8.0:
        return 0.0
    return float((-8.0 - sun_alt_deg) / 10.0)


def fading_score(hours_since_event: float, decay_rate: float = 0.08) -> float:
    # Exponential fading factor; intentionally mild so astronomy + visibility still dominate.
    return float(np.exp(-decay_rate * max(0.0, hours_since_event)))


def action_cost(telescope: Telescope, cost_weight: float = 0.02) -> float:
    block_hours = (telescope.exposure_s + telescope.overhead_s) / 3600.0
    return float(cost_weight * block_hours)


def covered_pixels_for_tile(tile: Tile, telescope: Telescope, nside: int) -> np.ndarray:
    center = SkyCoord(ra=tile.ra_deg * u.deg, dec=tile.dec_deg * u.deg, frame="icrs")
    vec = skycoord_to_vec(center)
    radius = np.radians(telescope.radius_deg)
    pix = hp.query_disc(nside, vec, radius, inclusive=False, nest=False)
    return np.asarray(pix, dtype=int)


def build_action_library(
    event: Dict[str, object],
    tiles: Sequence[Tile],
    telescopes: Sequence[Telescope],
    latency_hours: float,
    horizon_hours: float,
    slot_minutes: int,
    keep_top_actions_per_telescope: int,
) -> Dict[str, List[Action]]:
    prob = event["prob"]
    nside = int(event["nside"])
    event_time: Time = event["event_time"]
    slots = generate_time_slots(event_time, latency_hours, horizon_hours, slot_minutes)

    action_library: Dict[str, List[Action]] = {tel.name: [] for tel in telescopes}
    action_id_counter = 0

    for tel in telescopes:
        observer = tel.observer
        tile_pixel_cache: Dict[int, np.ndarray] = {}

        for slot_id, slot_time in enumerate(slots):
            hours_since_event = float((slot_time - event_time).to_value(u.hour))
            sun_alt = observer.altaz(slot_time, get_sun(slot_time)).alt.deg
            dark_score = darkness_score(float(sun_alt))
            if dark_score <= 0.0:
                continue

            for tile in tiles:
                coord = SkyCoord(ra=tile.ra_deg * u.deg, dec=tile.dec_deg * u.deg, frame="icrs")
                altaz = observer.altaz(slot_time, coord)
                alt_deg = float(altaz.alt.deg)
                vis_score = altitude_visibility_score(alt_deg, tel.min_alt_deg)
                if vis_score <= 0.0:
                    continue

                q_scalar = vis_score * dark_score * fading_score(hours_since_event) * tel.depth_weight
                q_scalar = float(np.clip(q_scalar, 0.0, 0.95))
                if q_scalar <= 0.02:
                    continue

                if tile.tile_id not in tile_pixel_cache:
                    tile_pixel_cache[tile.tile_id] = covered_pixels_for_tile(tile, tel, nside)
                covered = tile_pixel_cache[tile.tile_id]
                if covered.size == 0:
                    continue

                marginal_from_empty = q_scalar * float(prob[covered].sum()) - action_cost(tel)
                if marginal_from_empty <= 0:
                    continue

                action_library[tel.name].append(
                    Action(
                        action_id=action_id_counter,
                        telescope=tel.name,
                        tile_id=tile.tile_id,
                        slot_id=slot_id,
                        slot_time_iso=slot_time.isot,
                        ra_deg=tile.ra_deg,
                        dec_deg=tile.dec_deg,
                        radius_deg=tel.radius_deg,
                        q_scalar=q_scalar,
                        cost=action_cost(tel),
                        slot_hours_since_event=hours_since_event,
                        visibility_score=vis_score * dark_score,
                        depth_weight=tel.depth_weight,
                        covered_pixels=tuple(int(x) for x in covered.tolist()),
                    )
                )
                action_id_counter += 1

    # Diversified pruning: retain strong actions while preventing the pruned library
    # from collapsing onto one tile observed repeatedly at adjacent times.
    pruned: Dict[str, List[Action]] = {}
    for tel in telescopes:
        actions = action_library[tel.name]
        ranked = sorted(
            actions,
            key=lambda a: (a.q_scalar * float(prob[list(a.covered_pixels)].sum()) - a.cost, a.visibility_score, -a.slot_hours_since_event),
            reverse=True,
        )
        chosen: List[Action] = []
        per_tile_count: Dict[int, int] = {}
        per_slot_count: Dict[int, int] = {}
        for a in ranked:
            if len(chosen) >= keep_top_actions_per_telescope:
                break
            if per_tile_count.get(a.tile_id, 0) >= 2:
                continue
            if per_slot_count.get(a.slot_id, 0) >= 3:
                continue
            chosen.append(a)
            per_tile_count[a.tile_id] = per_tile_count.get(a.tile_id, 0) + 1
            per_slot_count[a.slot_id] = per_slot_count.get(a.slot_id, 0) + 1
        pruned[tel.name] = chosen

    return pruned


# -----------------------------------------------------------------------------
# Objective and schedule evaluation
# -----------------------------------------------------------------------------

def miss_vector_after_actions(prob: np.ndarray, n_pixels: int, actions: Sequence[Action]) -> np.ndarray:
    miss = np.ones(n_pixels, dtype=float)
    for action in actions:
        idx = np.fromiter(action.covered_pixels, dtype=int)
        miss[idx] *= (1.0 - action.q_scalar)
    return miss


def detection_probability_vector(n_pixels: int, actions: Sequence[Action]) -> np.ndarray:
    miss = np.ones(n_pixels, dtype=float)
    for action in actions:
        idx = np.fromiter(action.covered_pixels, dtype=int)
        miss[idx] *= (1.0 - action.q_scalar)
    return 1.0 - miss


def welfare_from_actions(prob: np.ndarray, actions: Sequence[Action]) -> float:
    det = detection_probability_vector(prob.size, actions)
    total_cost = float(sum(a.cost for a in actions))
    return float(np.sum(prob * det) - total_cost)


def expected_detection_score(prob: np.ndarray, actions: Sequence[Action]) -> float:
    det = detection_probability_vector(prob.size, actions)
    return float(np.sum(prob * det))


def binary_covered_mass(prob: np.ndarray, actions: Sequence[Action]) -> float:
    covered = np.zeros(prob.size, dtype=bool)
    for action in actions:
        idx = np.fromiter(action.covered_pixels, dtype=int)
        covered[idx] = True
    return float(prob[covered].sum())


def redundancy_mass(prob: np.ndarray, actions: Sequence[Action]) -> float:
    counts = np.zeros(prob.size, dtype=int)
    for action in actions:
        idx = np.fromiter(action.covered_pixels, dtype=int)
        counts[idx] += 1
    return float(prob[counts >= 2].sum())


def evaluate_subset_marginal_value(
    prob: np.ndarray,
    subset: Sequence[Action],
    miss_others: np.ndarray,
) -> float:
    miss = miss_others.copy()
    for action in subset:
        idx = np.fromiter(action.covered_pixels, dtype=int)
        miss[idx] *= (1.0 - action.q_scalar)
    gain = float(np.sum(prob * (miss_others - miss)))
    cost = float(sum(a.cost for a in subset))
    return gain - cost


# -----------------------------------------------------------------------------
# Feasible subset search
# -----------------------------------------------------------------------------

def is_feasible_subset(actions: Sequence[Action], max_actions: int) -> bool:
    if len(actions) > max_actions:
        return False
    slot_ids = [a.slot_id for a in actions]
    tile_ids = [a.tile_id for a in actions]
    return len(slot_ids) == len(set(slot_ids)) and len(tile_ids) == len(set(tile_ids))


def enumerate_feasible_subsets(actions: Sequence[Action], max_actions: int) -> Iterable[Tuple[Action, ...]]:
    yield tuple()
    max_r = min(max_actions, len(actions))
    for r in range(1, max_r + 1):
        for combo in itertools.combinations(actions, r):
            if is_feasible_subset(combo, max_actions):
                yield combo


def exact_best_response(
    prob: np.ndarray,
    candidate_actions: Sequence[Action],
    max_actions: int,
    miss_others: np.ndarray,
) -> Tuple[Tuple[Action, ...], float]:
    best_subset: Tuple[Action, ...] = tuple()
    best_value = evaluate_subset_marginal_value(prob, best_subset, miss_others)
    for subset in enumerate_feasible_subsets(candidate_actions, max_actions):
        value = evaluate_subset_marginal_value(prob, subset, miss_others)
        if value > best_value + 1e-12:
            best_value = value
            best_subset = subset
    return best_subset, best_value


# -----------------------------------------------------------------------------
# Schedulers / methods
# -----------------------------------------------------------------------------

def centralized_greedy_schedule(
    prob: np.ndarray,
    action_library: Dict[str, List[Action]],
    telescopes: Sequence[Telescope],
) -> Dict[str, Tuple[Action, ...]]:
    selected: Dict[str, List[Action]] = {tel.name: [] for tel in telescopes}
    used_slots: Dict[str, set] = {tel.name: set() for tel in telescopes}
    miss = np.ones(prob.size, dtype=float)

    action_pool = [a for actions in action_library.values() for a in actions]
    remaining = set(range(len(action_pool)))

    improved = True
    while improved:
        improved = False
        best_idx = None
        best_gain = 0.0

        for idx in list(remaining):
            a = action_pool[idx]
            tel = next(t for t in telescopes if t.name == a.telescope)
            if len(selected[a.telescope]) >= tel.max_actions:
                continue
            if a.slot_id in used_slots[a.telescope]:
                continue
            covered_idx = np.fromiter(a.covered_pixels, dtype=int)
            gain = float(np.sum(prob[covered_idx] * miss[covered_idx]) * a.q_scalar - a.cost)
            if gain > best_gain + 1e-12:
                best_gain = gain
                best_idx = idx

        if best_idx is not None and best_gain > 0:
            improved = True
            a = action_pool[best_idx]
            selected[a.telescope].append(a)
            used_slots[a.telescope].add(a.slot_id)
            covered_idx = np.fromiter(a.covered_pixels, dtype=int)
            miss[covered_idx] *= (1.0 - a.q_scalar)
            remaining.remove(best_idx)
        else:
            break

    return {k: tuple(v) for k, v in selected.items()}


def independent_greedy_schedule(
    prob: np.ndarray,
    action_library: Dict[str, List[Action]],
    telescopes: Sequence[Telescope],
) -> Dict[str, Tuple[Action, ...]]:
    """
    Selfish per-telescope scheduler: each telescope greedily maximizes its own
    standalone score without regard to overlap with other facilities.
    """
    schedule: Dict[str, Tuple[Action, ...]] = {}
    for tel in telescopes:
        chosen: List[Action] = []
        used_slots = set()
        used_tiles = set()
        miss = np.ones(prob.size, dtype=float)
        improved = True
        while improved and len(chosen) < tel.max_actions:
            improved = False
            best_action = None
            best_gain = 0.0
            for a in action_library[tel.name]:
                if a.slot_id in used_slots or a.tile_id in used_tiles:
                    continue
                idx = np.fromiter(a.covered_pixels, dtype=int)
                gain = float(np.sum(prob[idx] * miss[idx]) * a.q_scalar - a.cost)
                if gain > best_gain + 1e-12:
                    best_gain = gain
                    best_action = a
            if best_action is not None and best_gain > 0:
                improved = True
                chosen.append(best_action)
                used_slots.add(best_action.slot_id)
                used_tiles.add(best_action.tile_id)
                idx = np.fromiter(best_action.covered_pixels, dtype=int)
                miss[idx] *= (1.0 - best_action.q_scalar)
        schedule[tel.name] = tuple(chosen)
    return schedule


def random_feasible_schedule(
    action_library: Dict[str, List[Action]],
    telescopes: Sequence[Telescope],
    seed: int,
) -> Dict[str, Tuple[Action, ...]]:
    rng = random.Random(seed)
    schedule: Dict[str, Tuple[Action, ...]] = {}
    for tel in telescopes:
        actions = action_library[tel.name][:]
        rng.shuffle(actions)
        chosen: List[Action] = []
        used_slots = set()
        for a in actions:
            if len(chosen) >= tel.max_actions:
                break
            if a.slot_id in used_slots:
                continue
            chosen.append(a)
            used_slots.add(a.slot_id)
        schedule[tel.name] = tuple(chosen)
    return schedule


def proposed_potential_game_schedule(
    prob: np.ndarray,
    action_library: Dict[str, List[Action]],
    telescopes: Sequence[Telescope],
    max_rounds: int = 20,
) -> Tuple[Dict[str, Tuple[Action, ...]], List[Dict[str, object]]]:
    profile = independent_greedy_schedule(prob, action_library, telescopes)
    history: List[Dict[str, object]] = []

    for round_id in range(max_rounds):
        changed = False
        round_record = {"round": round_id + 1}
        for tel in telescopes:
            others = [a for k, acts in profile.items() if k != tel.name for a in acts]
            miss_others = miss_vector_after_actions(prob, prob.size, others)
            current_subset = profile[tel.name]
            current_value = evaluate_subset_marginal_value(prob, current_subset, miss_others)
            best_subset, best_value = exact_best_response(
                prob=prob,
                candidate_actions=action_library[tel.name],
                max_actions=tel.max_actions,
                miss_others=miss_others,
            )
            if best_value > current_value + 1e-12:
                profile[tel.name] = best_subset
                changed = True
                round_record[f"{tel.name}_changed"] = True
                round_record[f"{tel.name}_delta"] = float(best_value - current_value)
            else:
                round_record[f"{tel.name}_changed"] = False
                round_record[f"{tel.name}_delta"] = 0.0
        round_record["potential_after_round"] = welfare_from_actions(prob, [a for acts in profile.values() for a in acts])
        history.append(round_record)
        if not changed:
            break
    return profile, history


# -----------------------------------------------------------------------------
# Shapley credit on realized final schedule
# -----------------------------------------------------------------------------

def realized_schedule_shapley(
    prob: np.ndarray,
    schedule: Dict[str, Tuple[Action, ...]],
    telescope_names: Sequence[str],
) -> pd.DataFrame:
    telescopes = list(telescope_names)
    permutations = list(itertools.permutations(telescopes))
    contrib = {name: 0.0 for name in telescopes}

    for perm in permutations:
        coalition_actions: List[Action] = []
        coalition_value = welfare_from_actions(prob, coalition_actions)
        for name in perm:
            coalition_actions = coalition_actions + list(schedule.get(name, tuple()))
            new_value = welfare_from_actions(prob, coalition_actions)
            contrib[name] += new_value - coalition_value
            coalition_value = new_value

    for name in contrib:
        contrib[name] /= len(permutations)

    df = pd.DataFrame({
        "telescope": telescopes,
        "realized_shapley_value": [contrib[name] for name in telescopes],
        "n_actions": [len(schedule.get(name, tuple())) for name in telescopes],
    })
    return df.sort_values("realized_shapley_value", ascending=False).reset_index(drop=True)


# -----------------------------------------------------------------------------
# Plotting
# -----------------------------------------------------------------------------

def plot_skymap_with_schedule(
    prob: np.ndarray,
    nside: int,
    schedule: Dict[str, Tuple[Action, ...]],
    method_name: str,
    event_name: str,
    outpath: Path,
) -> None:
    cl = credible_levels(prob)
    pix = np.where(cl <= 0.90)[0]
    coords = hp_pix_to_skycoord(nside, pix)

    fig = plt.figure(figsize=(12, 6))
    ax = fig.add_subplot(111, projection="mollweide")
    sc = ax.scatter(
        wrap_ra_to_mollweide(coords.ra.deg),
        np.radians(coords.dec.deg),
        c=np.log10(prob[pix] + 1e-12),
        s=8,
        alpha=0.75,
        cmap="viridis",
        linewidths=0,
    )

    markers = ["o", "s", "^", "D", "P", "X"]
    for idx, (tel_name, actions) in enumerate(schedule.items()):
        if not actions:
            continue
        ras = np.array([a.ra_deg for a in actions])
        decs = np.array([a.dec_deg for a in actions])
        sizes = np.array([180 * a.radius_deg / 4.0 for a in actions])
        ax.scatter(
            wrap_ra_to_mollweide(ras),
            np.radians(decs),
            s=sizes,
            marker=markers[idx % len(markers)],
            edgecolor="black",
            linewidth=0.8,
            alpha=0.95,
            label=tel_name.replace("_", " "),
        )

    ax.grid(True, alpha=0.3)
    ax.set_title(f"{event_name}: 90% sky region with scheduled tiles\n{METHOD_LABELS[method_name]}", pad=16)
    cb = plt.colorbar(sc, ax=ax, shrink=0.75, pad=0.08)
    cb.set_label(r"log$_{10}$(pixel probability)")
    ax.legend(loc="lower left", fontsize=9)
    plt.tight_layout()
    fig.savefig(outpath, dpi=220, bbox_inches="tight")
    plt.show()
    plt.close(fig)


def plot_method_comparison(df_event_metrics: pd.DataFrame, event_name: str, outpath: Path) -> None:
    plot_df = df_event_metrics.copy()
    plot_df = plot_df.set_index("method").loc[METHOD_ORDER].reset_index()

    fig, axes = plt.subplots(2, 2, figsize=(12, 9))
    metrics = [
        ("expected_detection", "Expected detection"),
        ("binary_covered_mass", "Unique covered posterior mass"),
        ("redundancy_mass", "Overlap posterior mass"),
        ("welfare", "Net welfare"),
    ]

    for ax, (metric, label) in zip(axes.ravel(), metrics):
        ax.bar(
            [METHOD_LABELS[m] for m in plot_df["method"]],
            plot_df[metric],
            color=[METHOD_COLORS[m] for m in plot_df["method"]],
            alpha=0.9,
        )
        ax.set_title(label)
        ax.tick_params(axis="x", rotation=18)
        ax.grid(axis="y", alpha=0.25)

    fig.suptitle(f"Method comparison for {event_name}", fontsize=15)
    plt.tight_layout()
    fig.savefig(outpath, dpi=220, bbox_inches="tight")
    plt.show()
    plt.close(fig)


def plot_schedule_gantt(schedule_df: pd.DataFrame, event_name: str, method_name: str, outpath: Path) -> None:
    if schedule_df.empty:
        return
    tel_order = list(schedule_df["telescope"].drop_duplicates())
    y_map = {name: i for i, name in enumerate(tel_order)}

    fig, ax = plt.subplots(figsize=(12, 3 + 0.7 * len(tel_order)))
    for _, row in schedule_df.iterrows():
        ax.barh(
            y=y_map[row["telescope"]],
            width=row["block_duration_hours"],
            left=row["slot_hours_since_event"],
            height=0.55,
            alpha=0.9,
            color=METHOD_COLORS.get(method_name, "tab:blue"),
            edgecolor="black",
        )
        ax.text(
            row["slot_hours_since_event"] + 0.02,
            y_map[row["telescope"]],
            f"tile {int(row['tile_id'])}",
            va="center",
            ha="left",
            fontsize=8,
            color="black",
        )

    ax.set_yticks(list(y_map.values()))
    ax.set_yticklabels(tel_order)
    ax.set_xlabel("Hours since GW event")
    ax.set_title(f"{event_name}: schedule timeline ({METHOD_LABELS[method_name]})")
    ax.grid(axis="x", alpha=0.3)
    plt.tight_layout()
    fig.savefig(outpath, dpi=220, bbox_inches="tight")
    plt.show()
    plt.close(fig)


def plot_shapley(df_shapley: pd.DataFrame, event_name: str, outpath: Path) -> None:
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.bar(df_shapley["telescope"], df_shapley["realized_shapley_value"], color="tab:purple", alpha=0.9)
    ax.set_title(f"{event_name}: realized Shapley credit (proposed schedule)")
    ax.set_ylabel("Shapley value")
    ax.grid(axis="y", alpha=0.25)
    ax.tick_params(axis="x", rotation=18)
    plt.tight_layout()
    fig.savefig(outpath, dpi=220, bbox_inches="tight")
    plt.show()
    plt.close(fig)


def plot_aggregate_scores(df_all_metrics: pd.DataFrame, outpath: Path) -> None:
    agg = (
        df_all_metrics.groupby("method", as_index=False)[["expected_detection", "binary_covered_mass", "redundancy_mass", "welfare"]]
        .mean()
    )
    agg = agg.set_index("method").loc[METHOD_ORDER].reset_index()

    fig, axes = plt.subplots(1, 2, figsize=(11, 4.8))
    axes[0].bar(
        [METHOD_LABELS[m] for m in agg["method"]],
        agg["expected_detection"],
        color=[METHOD_COLORS[m] for m in agg["method"]],
        alpha=0.9,
    )
    axes[0].set_title("Mean expected detection across events")
    axes[0].tick_params(axis="x", rotation=18)
    axes[0].grid(axis="y", alpha=0.25)

    axes[1].bar(
        [METHOD_LABELS[m] for m in agg["method"]],
        agg["welfare"],
        color=[METHOD_COLORS[m] for m in agg["method"]],
        alpha=0.9,
    )
    axes[1].set_title("Mean welfare across events")
    axes[1].tick_params(axis="x", rotation=18)
    axes[1].grid(axis="y", alpha=0.25)

    plt.tight_layout()
    fig.savefig(outpath, dpi=220, bbox_inches="tight")
    plt.show()
    plt.close(fig)


# -----------------------------------------------------------------------------
# Tabulation
# -----------------------------------------------------------------------------

def schedule_to_dataframe(schedule: Dict[str, Tuple[Action, ...]], telescopes: Sequence[Telescope]) -> pd.DataFrame:
    tel_lookup = {t.name: t for t in telescopes}
    rows = []
    for tel_name, actions in schedule.items():
        tel = tel_lookup[tel_name]
        for a in actions:
            rows.append(
                {
                    "telescope": tel_name,
                    "slot_id": a.slot_id,
                    "slot_time_iso": a.slot_time_iso,
                    "slot_hours_since_event": a.slot_hours_since_event,
                    "tile_id": a.tile_id,
                    "ra_deg": a.ra_deg,
                    "dec_deg": a.dec_deg,
                    "radius_deg": a.radius_deg,
                    "q_scalar": a.q_scalar,
                    "cost": a.cost,
                    "visibility_score": a.visibility_score,
                    "block_duration_hours": (tel.exposure_s + tel.overhead_s) / 3600.0,
                    "n_covered_pixels": len(a.covered_pixels),
                }
            )
    df = pd.DataFrame(rows)
    if not df.empty:
        df = df.sort_values(["telescope", "slot_id", "tile_id"]).reset_index(drop=True)
    return df


def event_summary_dataframe(event: Dict[str, object]) -> pd.DataFrame:
    return pd.DataFrame(
        [
            {
                "event_name": event["event_name"],
                "gps": event["gps"],
                "event_time_utc": event["event_time"].isot if event["event_time"] is not None else None,
                "detectors": ", ".join(event["detectors"]),
                "nside_working": event["nside"],
                "pixel_area_sqdeg": event["pixel_area_sqdeg"],
                "area50_sqdeg": event["area50_sqdeg"],
                "area90_sqdeg": event["area90_sqdeg"],
                "area99_sqdeg": event["area99_sqdeg"],
                "skymap_url": event["skymap_url"],
            }
        ]
    )


def telescope_dataframe(telescopes: Sequence[Telescope]) -> pd.DataFrame:
    rows = []
    for t in telescopes:
        rows.append(
            {
                "telescope": t.name,
                "site_label": t.site_label,
                "lat_deg": t.lat_deg,
                "lon_deg": t.lon_deg,
                "height_m": t.height_m,
                "fov_sqdeg": t.fov_sqdeg,
                "radius_deg": t.radius_deg,
                "depth_weight": t.depth_weight,
                "exposure_s": t.exposure_s,
                "overhead_s": t.overhead_s,
                "max_actions": t.max_actions,
                "min_alt_deg": t.min_alt_deg,
            }
        )
    return pd.DataFrame(rows)


def action_library_dataframe(action_library: Dict[str, List[Action]], prob: np.ndarray) -> pd.DataFrame:
    rows = []
    for tel_name, actions in action_library.items():
        for a in actions:
            covered = np.fromiter(a.covered_pixels, dtype=int)
            rows.append(
                {
                    "telescope": tel_name,
                    "action_id": a.action_id,
                    "slot_id": a.slot_id,
                    "slot_time_iso": a.slot_time_iso,
                    "tile_id": a.tile_id,
                    "ra_deg": a.ra_deg,
                    "dec_deg": a.dec_deg,
                    "radius_deg": a.radius_deg,
                    "q_scalar": a.q_scalar,
                    "cost": a.cost,
                    "visibility_score": a.visibility_score,
                    "slot_hours_since_event": a.slot_hours_since_event,
                    "covered_posterior_mass": float(prob[covered].sum()),
                    "solo_net_gain": float(prob[covered].sum()) * a.q_scalar - a.cost,
                }
            )
    df = pd.DataFrame(rows)
    if not df.empty:
        df = df.sort_values(["telescope", "solo_net_gain"], ascending=[True, False]).reset_index(drop=True)
    return df


def metrics_dataframe(event_name: str, schedule_by_method: Dict[str, Dict[str, Tuple[Action, ...]]], prob: np.ndarray) -> pd.DataFrame:
    rows = []
    for method, schedule in schedule_by_method.items():
        actions = [a for acts in schedule.values() for a in acts]
        rows.append(
            {
                "event_name": event_name,
                "method": method,
                "n_actions": len(actions),
                "expected_detection": expected_detection_score(prob, actions),
                "binary_covered_mass": binary_covered_mass(prob, actions),
                "redundancy_mass": redundancy_mass(prob, actions),
                "total_cost": float(sum(a.cost for a in actions)),
                "welfare": welfare_from_actions(prob, actions),
                "n_telescopes_used": int(sum(1 for acts in schedule.values() if len(acts) > 0)),
                "n_unique_tiles": int(len(set(a.tile_id for a in actions))),
            }
        )
    return pd.DataFrame(rows)


# -----------------------------------------------------------------------------
# ZIP packaging
# -----------------------------------------------------------------------------

def zip_directory(directory: Path, zip_path: Path) -> None:
    with zipfile.ZipFile(zip_path, "w", compression=zipfile.ZIP_DEFLATED) as zf:
        for path in sorted(directory.rglob("*")):
            if path == zip_path:
                continue
            if path.is_file():
                zf.write(path, arcname=path.relative_to(directory))


# -----------------------------------------------------------------------------
# Main pipeline
# -----------------------------------------------------------------------------

def run_pipeline(args: argparse.Namespace) -> Dict[str, Path]:
    set_seed(args.seed)
    ensure_iers()

    base_output = Path(args.output_dir)
    safe_mkdir(base_output)
    cache_dir = base_output / "cache"
    safe_mkdir(cache_dir)

    telescopes = [Telescope(**cfg) for cfg in TELESCOPE_CONFIGS]

    all_metrics = []
    artifact_paths: Dict[str, Path] = {}

    tel_df = telescope_dataframe(telescopes)
    tel_df.to_csv(base_output / "telescope_network.csv", index=False)
    maybe_display_dataframe(tel_df, "Telescope network (illustrative but plausible)")

    for event_name in args.events:
        print("\n" + "#" * 120)
        print(f"Processing event: {event_name}")
        print("#" * 120)
        event_dir = base_output / event_name
        figs_dir = event_dir / "figures"
        tables_dir = event_dir / "tables"
        safe_mkdir(figs_dir)
        safe_mkdir(tables_dir)

        event = load_public_event_skymap(event_name, cache_dir, args.working_nside)
        prob = event["prob"]
        nside = int(event["nside"])

        event_df = event_summary_dataframe(event)
        event_df.to_csv(tables_dir / "event_summary.csv", index=False)
        maybe_display_dataframe(event_df, f"Event summary: {event_name}")

        representative_radius = float(np.median([t.radius_deg for t in telescopes]))
        min_sep = max(1.25, 0.95 * representative_radius)
        tiles = choose_tile_centers(
            prob=prob,
            nside=nside,
            max_tiles=args.max_tiles,
            candidate_credible_level=args.tile_candidate_credible_level,
            min_separation_deg=min_sep,
            selection_radius_deg=max(1.1, representative_radius),
        )
        tiles_df = pd.DataFrame([asdict(t) for t in tiles])
        tiles_df.to_csv(tables_dir / "candidate_tiles.csv", index=False)
        maybe_display_dataframe(tiles_df.head(30), f"Candidate tile centers: {event_name}")

        action_library = build_action_library(
            event=event,
            tiles=tiles,
            telescopes=telescopes,
            latency_hours=args.latency_hours,
            horizon_hours=args.horizon_hours,
            slot_minutes=args.slot_minutes,
            keep_top_actions_per_telescope=args.keep_top_actions_per_telescope,
        )
        action_df = action_library_dataframe(action_library, prob)
        action_df.to_csv(tables_dir / "pruned_action_library.csv", index=False)
        maybe_display_dataframe(action_df.head(60), f"Pruned action library: {event_name}")

        schedule_proposed, br_history = proposed_potential_game_schedule(prob, action_library, telescopes)
        schedule_central = centralized_greedy_schedule(prob, action_library, telescopes)
        schedule_independent = independent_greedy_schedule(prob, action_library, telescopes)
        schedule_random = random_feasible_schedule(action_library, telescopes, seed=args.seed + 17)

        schedule_by_method = {
            "proposed_potential_game": schedule_proposed,
            "centralized_greedy": schedule_central,
            "independent_greedy": schedule_independent,
            "random_feasible": schedule_random,
        }

        br_history_df = pd.DataFrame(br_history)
        br_history_df.to_csv(tables_dir / "best_response_history.csv", index=False)
        maybe_display_dataframe(br_history_df, f"Best-response history: {event_name}")

        metrics_df = metrics_dataframe(event_name, schedule_by_method, prob)
        metrics_df.to_csv(tables_dir / "method_metrics.csv", index=False)
        maybe_display_dataframe(metrics_df, f"Method comparison metrics: {event_name}")
        all_metrics.append(metrics_df)

        for method_name, schedule in schedule_by_method.items():
            sched_df = schedule_to_dataframe(schedule, telescopes)
            sched_df.to_csv(tables_dir / f"schedule_{method_name}.csv", index=False)
            maybe_display_dataframe(sched_df, f"Schedule table: {event_name} | {METHOD_LABELS[method_name]}")

            plot_skymap_with_schedule(
                prob=prob,
                nside=nside,
                schedule=schedule,
                method_name=method_name,
                event_name=event_name,
                outpath=figs_dir / f"skymap_{method_name}.png",
            )
            plot_schedule_gantt(
                schedule_df=sched_df,
                event_name=event_name,
                method_name=method_name,
                outpath=figs_dir / f"gantt_{method_name}.png",
            )

        plot_method_comparison(metrics_df, event_name, figs_dir / "method_comparison.png")

        shapley_df = realized_schedule_shapley(prob, schedule_proposed, [t.name for t in telescopes])
        shapley_df.to_csv(tables_dir / "realized_shapley_proposed.csv", index=False)
        maybe_display_dataframe(shapley_df, f"Realized Shapley credit: {event_name}")
        plot_shapley(shapley_df, event_name, figs_dir / "realized_shapley_proposed.png")

        zip_directory(event_dir, event_dir / f"{event_name}_results.zip")
        artifact_paths[f"{event_name}_results_zip"] = event_dir / f"{event_name}_results.zip"

    all_metrics_df = pd.concat(all_metrics, ignore_index=True)
    all_metrics_df.to_csv(base_output / "all_events_method_metrics.csv", index=False)
    maybe_display_dataframe(all_metrics_df, "All-event method metrics")
    plot_aggregate_scores(all_metrics_df, base_output / "aggregate_scores.png")

    # Lightweight narrative summary for the user after running.
    summary_lines = [
        "Pipeline completed successfully.",
        f"Events analysed: {', '.join(args.events)}",
        f"Working HEALPix nside: {args.working_nside}",
        f"Output directory: {base_output.resolve()}",
        "Benchmark methods: proposed potential game, centralized greedy, selfish independent greedy, random feasible.",
        "Interpretation reminder: the centralized greedy method is a strong planning reference, not a trivial baseline; the game-theoretic contribution is strongest when the decentralized potential-game schedule approaches centralized welfare while improving over selfish independent scheduling.",
    ]
    (base_output / "RUN_SUMMARY.txt").write_text("\n".join(summary_lines), encoding="utf-8")

    zip_directory(base_output, base_output / "gw_game_all_outputs.zip")
    artifact_paths["all_outputs_zip"] = base_output / "gw_game_all_outputs.zip"
    artifact_paths["output_dir"] = base_output
    return artifact_paths


# -----------------------------------------------------------------------------
# CLI
# -----------------------------------------------------------------------------

def build_arg_parser() -> argparse.ArgumentParser:
    parser = argparse.ArgumentParser(
        description="Game-theoretic GW follow-up pipeline using GWOSC / ligo.skymap / astroplan.",
        formatter_class=argparse.ArgumentDefaultsHelpFormatter,
    )
    parser.add_argument("--events", nargs="+", default=["GW170817", "GW190814"], help="Public events to analyse.")
    parser.add_argument("--output-dir", type=str, default="./gw_game_outputs", help="Directory for all saved results.")
    parser.add_argument("--working-nside", type=int, default=64, help="HEALPix nside used for computational work.")
    parser.add_argument("--latency-hours", type=float, default=2.0, help="Response latency after the GW event.")
    parser.add_argument("--horizon-hours", type=float, default=12.0, help="Scheduling horizon after latency.")
    parser.add_argument("--slot-minutes", type=int, default=30, help="Time-slot length in minutes.")
    parser.add_argument("--max-tiles", type=int, default=18, help="Maximum number of candidate tile centers.")
    parser.add_argument("--tile-candidate-credible-level", type=float, default=0.99, help="Credible region used to seed candidate tiles.")
    parser.add_argument("--keep-top-actions-per-telescope", type=int, default=18, help="Pruned action count per telescope for exact local search.")
    parser.add_argument("--seed", type=int, default=123, help="Random seed for reproducibility.")
    return parser


def main() -> None:
    parser = build_arg_parser()
    args, _ = parser.parse_known_args()
    artifact_paths = run_pipeline(args)
    print("\nArtifacts created:")
    for key, value in artifact_paths.items():
        print(f"- {key}: {value}")


if __name__ == "__main__":
    main()
```
